```python
import panorama
import imutils
import cv2
import matplotlib.pyplot as plt

# Definir las rutas de las imágenes directamente
image_paths = [
    "pictures/foto (1).jpeg",
    "pictures/foto (2).jpeg",
    "pictures/foto (3).jpeg",
    "pictures/foto (4).jpeg",
    "pictures/foto (5).jpeg",
    "pictures/foto (6).jpeg",
    "pictures/foto (7).jpeg",
    "pictures/foto (8).jpeg",
    "pictures/foto (9).jpeg",
    "pictures/foto (10).jpeg",
    "pictures/foto (11).jpeg",
    "pictures/foto (12).jpeg",
    "pictures/foto (13).jpeg",
    "pictures/foto (14).jpeg",
    # Agrega más rutas si es necesario
]

images = []
for path in image_paths:
    images.append(cv2.imread(path))

for i in range(len(images)):
    images[i] = imutils.resize(images[i], width=100)
    images[i] = imutils.resize(images[i], height=100)

if len(images) == 2:
    (result, matched_points) = panorama.image_stitch([images[0], images[1]], match_status=True)
else:
    (result, matched_points) = panorama.image_stitch([images[-2], images[-1]], match_status=True)
    for i in range(len(images) - 2):
        (result, matched_points) = panorama.image_stitch([images[-i - 3], result], match_status=True)

plt.imshow(matched_points)
plt.show()
plt.imshow(result)
plt.show()
```
